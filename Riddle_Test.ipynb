```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import seaborn as sns
sns.set(rc={'figure.figsize':(30,10)})
```

Імпортуємо дані з наданої таблиці

```python
df=pd.read_csv("/content/Тестовое_аналитик.xlsx - Данные из таблицы.csv",decimal=',')
df.head()
```

Так як перший рядок має максимальні значення і може завадити обробці данних, вирішив його видалити


```python

df.drop(labels=0, inplace=True)
df.head()
```

Дані у таблиці не у хронологічному порядку, вирішив прослідкувати зміну параметрів день за днем

```python
df.sort_values(by=['Дата'], inplace=True)
df.head(5)
```

Так як у звичайних людей 5 денний робочий тиждень, вирішив додати стовбчик " days", щоб розуміти у який день тижня більше всього активних переглядів. 

```python
df['Day'] = pd.DatetimeIndex(df['Дата']).day_name()
df.head(10)
```

Підготувавши таблицю, використав бібліотеку Seaborn для побудови графіку залежності змінних одна від одної. Це дає змогу будувати гіпотези без додаткових робіт з таблицею.

```python
cor = df.corr()
sns.heatmap(data=cor, annot=True)
```

З данної " Heatmap" можна побачити, що наприклад кількість Дизлайків ніяк не корелює з опублікованими відео. Це дає нам  змогу зрозуміти що дизлайки не ставлять тільки аби знизити рейтинг відео, чи каналу. Інший приклад Підписники та перегляди, коеф. 0.87 говорить про те, що майже уся частина переглядів припадає на підписників.

Вирішив подивитится який розподіл переглядів. З наведеного нижче графіку видно що здебільшого це 30 тисяч щодня 

```python
plot = sns.distplot(df["Просмотры"])
plt.show()
```

```python
fig, ax1 = plt.subplots(2)

x=df['Day']
y=df['Отметки "Нравится"']
z=df['Новые подписчики']
q=df["Просмотры"]
d=df["Доход на тысячу просмотров (USD)"]

ax1[0].bar(x, q, color='green',alpha=0.6, label =("Просмотры по дням"))
ax1[1].bar(x, z, color='red',alpha=0.2, label =("Новые подписчики по дням"))
ax1[0].legend()
ax1[1].legend()
```

Графіки вище дозволяють нам зрозуміти що найактивніше перегляди та нові підписники приєднуються у субботу та неділю. А це в свою чергу значить що дні по яким краще викладати відео це п'ятниця/субота.

```python
ax2=sns.barplot(x=d,y=x, data=df)
```

А тут у свою чергу провів аналіз, який дає контрінтуїтивні висновки. З графіку можна вирішити що краще за все "Доход на тысячу просмотров (USD)" по понеділкам та середам. Однак звернувши увагу на шкалу увесь прибуток знаходиться у діапазоні 5-6. А це значить, що змінивши дату публікації відео, в рази більший дохід отримано не буде.
