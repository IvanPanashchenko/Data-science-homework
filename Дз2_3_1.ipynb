```python
data.groupby('genre')['price'].agg(['min', 'max'])

```

```python
second_df = data.groupby('author')['name'].agg(['count'])
second_df.shape

```

```python
second_df[second_df['count'] == second_df['count'].max()]

```

```python
third_df = data.groupby('author')['user_rating'].agg(['mean']) 
third_df[third_df['mean'] == second_df['mean'].min()]

```

```python
inter_data = pd.concat([second_df, third_df], axis = 1)
ax=inter_data.sort_values(by = ['count', 'mean'])
```

```python
ay=sns.barplot(x="count",y="mean", data=ax)
```

```python
sns.lineplot(x="reviews",y="user_rating", data=az, hue="user_rating")
```

```python
sns.displot(x="reviews", data=az, color="orange")
```
